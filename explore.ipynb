```python
from bs4 import BeautifulSoup, Comment
import pandas as pd
import os
from io import StringIO

file_paths = ["data/raw/html/Arsenal Goal Logs, Champions League _ FBref.com.html",
              'data/raw/html/Arsenal Goal Logs, Premier League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Goalkeeping), Champions League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Goalkeeping), Premier League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Miscellaneous Stats), Champions League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Miscellaneous Stats), Premier League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Shooting), Champions League _ FBref.com.html',
              'data/raw/html/Arsenal Match Logs (Shooting), Premier League _ FBref.com.html',
              'data/raw/html/Arsenal Scores and Fixtures, Premier League _ FBref.com.html',
              'data/raw/html/Arsenal Scores and Fixtures, Champions League _ FBref.com.html',
              ]


for file_path in file_paths:
    with open(file_path, "r", encoding="utf-8") as f:
        html = f.read()
    print("=" * 50)

    # CHECK RAW HTML
    print("Length of HTML:", len(html))
    print("Number of <table tags:", html.lower().count("<table"))

    print("=" * 50)

    soup = BeautifulSoup(html, "html.parser")

    # NORMAL TABLES
    tables = soup.find_all("table")
    print("BeautifulSoup visible tables:", len(tables))

    print("=" * 50)

    # COMMENT TABLES
    comments = soup.find_all(string=lambda text: isinstance(text, Comment))

    comment_tables = 0

    for comment in comments:
        if "<table" in comment.lower():
            comment_tables += 1

    print("Tables hidden in comments:", comment_tables)

    print("=" * 50)
    
    folder_name = os.path.splitext(os.path.basename(file_path))[0]
    folder_name = folder_name.lower().replace(" ", "_")
            
    output_dir = f"debug_tables/{folder_name}"

    # TRY PANDAS
    try:
        dfs = pd.read_html(StringIO(html))
        print("Pandas found tables:", len(dfs))

        import os

        os.makedirs(output_dir, exist_ok=True)

        for i, df in enumerate(dfs):

            csv_path = f"{output_dir}/table_{i}.csv"

            df.to_csv(csv_path, index=False)

            print(f"Saved {csv_path}")

    except Exception as e:
        print("Pandas failed:")
        print(e)

    print("=" * 50)
```

```python
import os
import glob

base_dir = "debug_tables"

for folder in os.listdir(base_dir):

    folder_path = os.path.join(base_dir, folder)

    if os.path.isdir(folder_path):

        csv_files = sorted(
            glob.glob(f"{folder_path}/*.csv"),
            key=lambda x: int(
                os.path.basename(x)
                .replace("table_", "")
                .replace(".csv", "")
            )
        )

        # keep last 2
        files_to_delete = csv_files[:-2]

        for file in files_to_delete:
            os.remove(file)
            print("Deleted:", file)
```
